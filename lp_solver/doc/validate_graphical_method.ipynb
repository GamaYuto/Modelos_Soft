# Validación de graphical_method.py

Este notebook valida el archivo `graphical_method.py` verificando:
1. Sintaxis Python válida
2. Importación correcta del módulo
3. Ejecución de los 3 tests unitarios
4. Reporte final de resultados

## 1. Validar Sintaxis del Archivo Python

Utilizaremos el módulo `py_compile` para compilar y validar que no hay errores de sintaxis.

```python
import py_compile
import sys
import os

# Cambiar al directorio del backend
os.chdir('/workspaces/Modelos_Soft/lp_solver/backend')
sys.path.insert(0, '.')

file_path = 'graphical_method.py'

print(f"Validando sintaxis de: {file_path}")
print("=" * 70)

try:
    py_compile.compile(file_path, doraise=True)
    print(f"✓ El archivo {file_path} tiene sintaxis Python válida")
    sintaxis_valida = True
except py_compile.PyCompileError as e:
    print(f"✗ Error de sintaxis en {file_path}:")
    print(e)
    sintaxis_valida = False

print("=" * 70)
print()
```

## 2. Importar el Módulo graphical_method

Intentaremos importar el módulo para identificar cualquier error de importación.

```python
print("Intentando importar el módulo graphical_method...")
print("=" * 70)

try:
    import graphical_method
    from graphical_method import resolver_grafico
    print("✓ Módulo 'graphical_method' importado exitosamente")
    print(f"✓ Función 'resolver_grafico' disponible")
    importacion_exitosa = True
except Exception as e:
    print(f"✗ Error al importar: {e}")
    import traceback
    traceback.print_exc()
    importacion_exitosa = False

print("=" * 70)
print()
```

## 3. Ejecutar el Archivo de Pruebas

Ejecutaremos los 3 tests definidos en `test_graphical.py` y capturaremos los resultados.

```python
print("EJECUTANDO TESTS DEL ARCHIVO test_graphical.py")
print("=" * 70)
print()

# Almacenar resultados
resultados_tests = []

# Test 1: Problema simple de maximización
print("=" * 70)
print("Test 1: Maximización simple")
print("=" * 70)
tipo = "max"
c = [3, 5]
restricciones = [
    [1, 0, "<=", 4],      # x1 <= 4
    [0, 2, "<=", 12],     # 2*x2 <= 12
    [3, 2, "<=", 18],     # 3*x1 + 2*x2 <= 18
]

try:
    resultado1 = resolver_grafico(tipo, c, restricciones)
    print(f"Status:        {resultado1['status']}")
    print(f"Valor óptimo:  {resultado1['valor_optimo']}")
    print(f"Conclusión:    {resultado1['conclusion']}")
    print("✓ Test 1 completado exitosamente")
    resultados_tests.append({
        'test': 1,
        'nombre': 'Maximización simple',
        'status': resultado1['status'],
        'valor_optimo': resultado1['valor_optimo'],
        'conclusion': resultado1['conclusion'],
        'exitoso': True
    })
except Exception as e:
    print(f"✗ Error en Test 1: {e}")
    import traceback
    traceback.print_exc()
    resultados_tests.append({
        'test': 1,
        'nombre': 'Maximización simple',
        'error': str(e),
        'exitoso': False
    })
print()

# Test 2: Problema infactible
print("=" * 70)
print("Test 2: Problema infactible")
print("=" * 70)
tipo = "max"
c = [1, 1]
restricciones = [
    [1, 1, "<=", 1],
    [1, 1, ">=", 2],
]

try:
    resultado2 = resolver_grafico(tipo, c, restricciones)
    print(f"Status:        {resultado2['status']}")
    print(f"Conclusión:    {resultado2['conclusion']}")
    print("✓ Test 2 completado exitosamente")
    resultados_tests.append({
        'test': 2,
        'nombre': 'Problema infactible',
        'status': resultado2['status'],
        'valor_optimo': resultado2.get('valor_optimo', 'N/A'),
        'conclusion': resultado2['conclusion'],
        'exitoso': True
    })
except Exception as e:
    print(f"✗ Error en Test 2: {e}")
    import traceback
    traceback.print_exc()
    resultados_tests.append({
        'test': 2,
        'nombre': 'Problema infactible',
        'error': str(e),
        'exitoso': False
    })
print()

# Test 3: Múltiples soluciones
print("=" * 70)
print("Test 3: Múltiples soluciones")
print("=" * 70)
tipo = "max"
c = [1, 1]
restricciones = [
    [1, 1, "<=", 1],
    [1, 0, ">=", 0],
    [0, 1, ">=", 0],
]

try:
    resultado3 = resolver_grafico(tipo, c, restricciones)
    print(f"Status:               {resultado3['status']}")
    print(f"Valor óptimo:        {resultado3['valor_optimo']}")
    print(f"Múltiples soluciones: {resultado3['multiple_solutions']}")
    print(f"Conclusión:           {resultado3['conclusion']}")
    print("✓ Test 3 completado exitosamente")
    resultados_tests.append({
        'test': 3,
        'nombre': 'Múltiples soluciones',
        'status': resultado3['status'],
        'valor_optimo': resultado3['valor_optimo'],
        'multiple_solutions': resultado3['multiple_solutions'],
        'conclusion': resultado3['conclusion'],
        'exitoso': True
    })
except Exception as e:
    print(f"✗ Error en Test 3: {e}")
    import traceback
    traceback.print_exc()
    resultados_tests.append({
        'test': 3,
        'nombre': 'Múltiples soluciones',
        'error': str(e),
        'exitoso': False
    })
print()
```

## 4. Analizar Resultados de los Tests

Resumen de los resultados obtenidos en cada test.

```python
print("=" * 70)
print("ANÁLISIS DE RESULTADOS")
print("=" * 70)
print()

# Crear tabla de resultados
print("Resumen de Tests Ejecutados:")
print("-" * 70)

for resultado in resultados_tests:
    print()
    print(f"Test {resultado['test']}: {resultado['nombre']}")
    if resultado['exitoso']:
        print(f"  Status:       {resultado['status']}")
        print(f"  Valor óptimo: {resultado['valor_optimo']}")
        print(f"  Conclusión:   {resultado['conclusion'][:80]}...")
    else:
        print(f"  Error: {resultado['error']}")
    print()

# Contar resultados
tests_exitosos = sum(1 for r in resultados_tests if r['exitoso'])
tests_fallidos = sum(1 for r in resultados_tests if not r['exitoso'])

print("=" * 70)
print(f"Total de Tests: {len(resultados_tests)}")
print(f"Tests Exitosos: {tests_exitosos}")
print(f"Tests Fallidos: {tests_fallidos}")
print("=" * 70)
print()
```

## 5. Generar Reporte Final

Reporte completo de validación de `graphical_method.py`

```python
print("╔" + "=" * 68 + "╗")
print("║" + " " * 68 + "║")
print("║" + "REPORTE FINAL DE VALIDACIÓN".center(68) + "║")
print("║" + " " * 68 + "║")
print("╚" + "=" * 68 + "╝")
print()

# Sección 1: Errores de Sintaxis
print("1. VALIDACIÓN DE SINTAXIS PYTHON:")
print("-" * 70)
if sintaxis_valida:
    print("   ✓ NO HAY ERRORES DE SINTAXIS")
    print("   El archivo graphical_method.py tiene una sintaxis Python válida")
else:
    print("   ✗ HAY ERRORES DE SINTAXIS")
print()

# Sección 2: Importación
print("2. IMPORTACIÓN DEL MÓDULO:")
print("-" * 70)
if importacion_exitosa:
    print("   ✓ MÓDULO IMPORTADO EXITOSAMENTE")
    print("   - graphical_method fue importado sin problemas")
    print("   - Función resolver_grafico() está disponible")
else:
    print("   ✗ ERROR EN LA IMPORTACIÓN")
print()

# Sección 3: Resultados de Tests
print("3. RESULTADOS DE LOS 3 TESTS:")
print("-" * 70)
for resultado in resultados_tests:
    if resultado['exitoso']:
        print(f"\n   Test {resultado['test']}: {resultado['nombre']}")
        print(f"   Status:       ✓ {resultado['status']}")
        print(f"   Valor óptimo: {resultado['valor_optimo']}")
        if resultado['test'] == 3:
            print(f"   Múltiples:    {resultado['multiple_solutions']}")
    else:
        print(f"\n   Test {resultado['test']}: {resultado['nombre']}")
        print(f"   Error:        ✗ {resultado['error']}")
print()

# Sección 4: Conclusión
print("4. CONCLUSIÓN FINAL:")
print("-" * 70)

if sintaxis_valida and importacion_exitosa and tests_exitosos == 3:
    print("   ✓✓✓ TODO FUNCIONA CORRECTAMENTE ✓✓✓")
    print()
    print("   Resumen:")
    print("   • Sintaxis: Válida")
    print("   • Importación: Exitosa")
    print("   • Tests: Los 3 tests ejecutados con éxito")
    print("   • Función resolver_grafico(): Operacional")
    print()
    print("   Status General: ✓ APROBADO")
else:
    print("   ✗ HAY PROBLEMAS A RESOLVER")
    if not sintaxis_valida:
        print("   • Error de sintaxis detectado")
    if not importacion_exitosa:
        print("   • Error de importación detectado")
    if tests_exitosos < 3:
        print(f"   • {tests_fallidos} test(s) fallido(s)")
    print()
    print("   Status General: ✗ RECHAZADO")

print()
print("╔" + "=" * 68 + "╗")
print("║" + "FIN DEL REPORTE".center(68) + "║")
print("╚" + "=" * 68 + "╝")
```
